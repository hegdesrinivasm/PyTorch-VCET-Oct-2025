### Reshaping:
Tensors can be reshaped using the 'reshape' method -- but only if the total number of elements remains the same. 

```python
x.reshape(9, 3)
```

```python
a = np.array([[[1,2], [1, 2]], [[1, 2], [1, 2]]])
a = torch.tensor(a)
a.shape
```

```python
a.reshape(4, 2)
```
